#### Imports

```python
import sys
sys.path.insert(0, '../package')

from package import *

def scale_data(train, test=''):    # Also adding scaling function.
    scaler = StandardScaler()

    train_data = scaler.fit_transform(train)
    
    if type(test) != str:
        test_data = scaler.transform(test)
        return train_data, test_data
    else:
        return train_data

display(dataset)
```

#### Prepare data for model

```python
target = (dataset['Playoffs'])
data_all = (dataset[['OBP', 'SLG', 'BA', 'ERA', 'ERA+', 'FIP', 'Ks', 'HR', 'RBIs', 'RC']])
features = ['OBP', 'SLG', 'BA', 'ERA', 'ERA+', 'FIP', 'Ks', 'HR', 'RBIs', 'RC']

display(target)
display(data_all)

# ---- VARIABLE DESCRIPTIONS ----
# OBP:          on-base percentage
# SLG:          slugging percentage
# BA:           batting average
# Playoffs:     binary (if team made playoffs or not)
# ERA:          earned run average
# ERA+:         earned run average +
# FIP:          fielding independent pitching
# HR:           home runs
# RBIs:         runs batted in
# RC:           runs created
# Ks:           strikeouts

target = target.to_numpy()
data_all = data_all.to_numpy()
```

#### Split up data

```python
X_train, X_test, y_train, y_test = train_test_split(data_all, target, stratify=target)

X_train, X_test = scale_data(X_train, X_test)

print(f'X_train.shape: {X_train.shape}; y_train.shape: {y_train.shape}')
print(f'X_test.shape: {X_test.shape}; y_test.shape: {y_test.shape}')
```

#### Create model

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(max_depth=2, learning_rate=0.05)

xgb.fit(X_train, y_train)

print(f'Training Accuracy: {xgb.score(X_train, y_train)*100:.2f}%')
print(f'Testing Accuracy: {xgb.score(X_test, y_test)*100:.2f}%')

# Even when fitting a more advanced statistical model... this is an optimal accuracy without the model overfitting.
```

#### Test model for different eras

```python
def time_period_test(year, symbol):
    if symbol == '<':
        dataset_time_period = dataset.loc[dataset['Year'] < year]
    elif symbol == '>':
        dataset_time_period = dataset.loc[dataset['Year'] > year]
    else:
        print(f"Invalid symbol '{symbol}'")

    data_time_period = (dataset_time_period[features]).to_numpy()
    target_time_period = (dataset_time_period['Playoffs']).to_numpy()

    data_time_period = data_time_period

    X_train, X_test, y_train, y_test = train_test_split(data_time_period, target_time_period)

    X_train, X_test = scale_data(X_train, X_test)
    
    model = XGBClassifier(max_depth=2, learning_rate=0.05)
    model.fit(X_train, y_train)

    accuracy_train = model.score(X_train, y_train)*100
    accuracy_test = model.score(X_test, y_test)*100

    fig = plt.figure(figsize=(10, 8))
    
    title = fig.suptitle(f'Statistics from {symbol} {year}\n Train accuracy: {accuracy_train:.1f}% \n Test accuracy: {accuracy_test:.1f}%')
    ax = fig.add_subplot(111, projection='3d')

    xs = data_time_period[:, 0]    # SLG
    ys = data_time_period[:, 1]    # OBP
    zs = data_time_period[:, 9]    # RC
    data_points = [(x, y, z) for x, y, z in zip(xs, ys, zs)]
    colors = ['green' if tar == 1 else 'red' for tar in target_time_period]

    for data, color in zip(data_points, colors):
        x, y, z = data
        ax.scatter(x, y, z, alpha=0.4, c=color, edgecolors='none', s=30)
    
    ax.set_xlabel('SLG')
    ax.set_ylabel('OBP')
    ax.set_zlabel('FIP')
```

```python
time_period_test(1993, '<')
```

```python
time_period_test(1994, '>')    # Significant increase in accuracy from linear model.
```

#### Make manual test to see if a team would make the playoffs

```python
def manual_test(team, year):
    team_object = Teams(year)[team]

    team_OBP = team_object.on_base_percentage
    team_SLG = team_object.slugging_percentage
    team_BA = team_object.batting_average
    team_ERA = team_object.earned_runs_against
    team_ERA_plus = team_object.earned_runs_against_plus
    team_FIP = team_object.fielding_independent_pitching
    team_Ks = team_object.strikeouts
    team_HR = team_object.home_runs
    team_RBIs = team_object.runs_batted_in
    team_RC = (team_object.hits + team_object.bases_on_balls) * team_object.total_bases / (team_object.at_bats + team_object.bases_on_balls)

    sample = [team_OBP, team_SLG, team_BA, team_ERA, team_ERA_plus, team_FIP, team_Ks, team_HR, team_RBIs, team_RC]
    sample = np.array(sample).reshape(1, -1)
    _, sample = scale_data(data_all, sample)    # Reinitializing scaler to scale sample.

    return f'Model prediction for {year} {team_object.abbreviation}: {xgb.predict(sample)[0]}'

for year in [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2021]:    # Skip 2020 (COVID year with 60 games)
    print(manual_test('CHC', year))

# NOTE: we don't have to include this in paper... just for moneyball sim... because of weird results.
```
